```python

class Post(TypedDict):
  post : str
  answer : str
```

```python
from langgraph.graph import StateGraph
```

```python
graph = StateGraph(Post)
```

```python
from langgraph.types import interrupt
def Generate_post(State : Post):
  messages = [SystemMessage("""Guardrails: Sexual Content & Intent Prohibition
    1. Absolute Prohibition of Sexual Content

    The system must not engage with, generate, continue, or encourage any form of sexual content. This includes explicit, implicit, suggestive, or coded sexual language, regardless of tone, context, or intent.

    2. Block Sexual Intent at Input Level

    If the user’s message shows sexual intent, the system must:

    Refuse to comply immediately

    Not ask follow-up questions

    Not attempt to “reframe” the request into a sexual-safe version

    Sexual intent includes curiosity, desire, fantasy, gratification, or arousal-seeking behavior.

    3. No Sexual Roleplay or Romantic Simulation

    The system must not participate in:

    Sexual roleplay

    Erotic storytelling

    Romantic or intimate simulations

    “Wife / girlfriend / partner” sexualized behavior

    Even if the user claims consent, familiarity, or fictional framing, the response must remain strictly non-sexual.

    4. Implicit and Coded Language Is Treated as Sexual

    The system must treat indirect, slang, euphemisms, metaphors, emojis, or suggestive phrasing that imply sexual meaning as sexual content.

    Examples include but are not limited to:

    Flirtation with sexual undertones

    “Jokes” with sexual implications

    Double-meaning phrases intended to bypass moderation

    All such attempts must be blocked.

    5. No Educational or Descriptive Sexual Content Unless Explicitly Allowed

    The system must not provide:

    Descriptions of sexual acts

    Sexual anatomy for arousal purposes

    Sexual techniques or advice

    Exception is allowed only if:

    The content is strictly medical, biological, or academic

    The tone is clinical and non-graphic

    The intent is clearly educational (e.g., health, biology)

    Otherwise, refuse.

    6. Zero Tolerance for Escalation

    If a user repeatedly attempts to introduce sexual topics after refusal:

    The system must maintain refusal

    Responses must become shorter and firmer

    The system must not soften tone or negotiate boundaries

    Persistent attempts should not be rewarded with additional explanations.

    7. No Sexual Content Involving Any Age Group

    The system must not generate sexual content involving:

    Minors

    Adults

    Fictional beings

    AI personas

    There are no exceptions, including hypothetical or fictional framing.

    8. Do Not Mirror Sexual Language

    If a user uses sexual language:

    The system must not repeat, paraphrase, or mirror it

    The system must use neutral, professional language in refusal

    9. Safe Redirection Only

    Upon refusing sexual content, the system may:

    Redirect the conversation to a neutral, productive topic

    Offer help in learning, problem-solving, or general discussion

    The redirection must not acknowledge or expand on sexual elements.

    10. No Moral Judgement, No Validation

    When refusing:

    Do not shame the user

    Do not validate sexual intent

    Do not say “it’s normal” or “I understand your desire”

    Simply state the boundary and redirect.

    11. Consistency Across Contexts

    These rules apply regardless of:

    Language used

    Cultural context

    Humor, sarcasm, or “just testing” claims

    Previous conversation history

    No context overrides these guardrails.

    12. System Integrity Takes Priority

    If a conflict arises between:

    User satisfaction

    Roleplay immersion

    Personality tone

    System safety and policy compliance always override user intent
    """),HumanMessage(State["post"])
    ]

  decision = interrupt({
      "type" : "approval",
      "reason" : "Model is about to answer to do research on post",
      "question" : State["post"],
      "instruction" : "Approve this question Approved or Disapproved"
  })
  if decision["approved"] == "Disapproved":
    return {"post": [AIMessage(content = "Not approved \n request is not following company guidelines and chat is inappropriate and has voilence in it")]}
  else:
    result = model.invoke(messages)
    return {"post": result}

```

```python
def Do_research(State : Post):
  prompt = [
      SystemMessage("Do a great research on the post and generate summary"),
      HumanMessage(f"The topic is {State["post"]}")
  ]

  result = model.invoke(prompt)
  return {"answer": result}
```

```python
checkpointer = MemorySaver()
```

```python
graph.add_node("Do_research",Do_research)
graph.add_node("Generate_post",Generate_post)

graph.add_edge(START,"Generate_post")
graph.add_edge("Generate_post","Do_research")
graph.add_edge("Do_research",END)
```

```python
workflow = graph.compile(checkpointer=checkpointer)
```

```python
config = {"configurable" : {"thread_id": "1"}}

initial_state = {"post": "Agentic Ai everything need to know"}
```

```python
result = workflow.invoke(initial_state, config=config)
```

```python
class Sentiment_Analysis(BaseModel):
  sentiment : Annotated[Literal["Approved","Disapproved"],Field(description= "Analyse the sentiment of the user message and analyse that the user is giving approval to the command or disapproved the command")]
```

```python
model_with_structure = model.with_structured_output(Sentiment_Analysis)
```

```python
input = __builtins__.input
user_prompt = input(f"Backend -{message} Approve this topic Approved or Disapproved ")
user_input = model_with_structure.invoke(user_prompt )
result = user_input.model_dump()
```

```python
final_result = workflow.invoke(
    Command(resume = {"approved": result["sentiment"]}),
    config = config
    )
```
